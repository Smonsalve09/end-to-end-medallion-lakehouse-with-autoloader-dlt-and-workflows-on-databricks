
# Bronze Zone


## 1. Esquema y tablas

Crear un esquema llamado **“bronze"** en el catálogo **“workspace”** y crear tablas con los datos de cada archivo definiendo la metadata. 

Nombrar tablas como:

* raw_empleados
* raw_locales
* raw_productos
* raw_fact


**SCHEMA** **BRONZE**

```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.bronze;
```

```python
from pyspark.sql.functions import *
```

**EMPLEADOS**

```python
empleados_df=spark.read.format("csv").option("header","true").option("inferSchema","true").load("/Volumes/workspace/default/test_volume/empleados.csv").withColumn("ingestion_date", current_timestamp()).withColumn("source_file", col("_metadata.file_path"))
empleados_df.printSchema()
```

```python
empleados_df.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.bronze.raw_empleados"
    )
```

**FACT**

```python
fact_df=spark.read.format("csv").option("header","true").option("delimiter",";").option("inferSchema","true").load("/Volumes/workspace/default/test_volume/fact.csv").withColumn("ingestion_date", current_timestamp()).withColumn("source_file", col("_metadata.file_path"))
fact_df.printSchema()
```

```python
fact_df.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.bronze.raw_fact"
    )
```

**LOCALES**

```python
locales_df=spark.read.format("csv").option("header","true").option("delimiter",";").option("inferSchema","true").load("/Volumes/workspace/default/test_volume/locales.csv").withColumn("ingestion_date", current_timestamp()).withColumn("source_file", col("_metadata.file_path"))
locales_df.printSchema()
```

```python
locales_df.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.bronze.raw_locales"
    )
```

**PRODUCTO**

```python
producto_df=spark.read.format("csv").option("header","true").option("delimiter",";").option("inferSchema","true").load("/Volumes/workspace/default/test_volume/producto.csv").withColumn("ingestion_date", current_timestamp()).withColumn("source_file", col("_metadata.file_path"))
producto_df.printSchema()
```

```python
producto_df.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.bronze.raw_producto"
    )
```


## 2. Validación capa bronze

Validar con una consulta que se hayan cargado correctamente todos los registros de cada uno de los archivos.

```python
%sql
SELECT
(SELECT COUNT(*) FROM workspace.bronze.raw_empleados) AS EMPLEADOS,
(SELECT COUNT(*) FROM workspace.bronze.raw_fact) AS VENTAS,
(SELECT COUNT(*) FROM workspace.bronze.raw_producto) AS PRODUCTOS,
(SELECT COUNT(*) FROM workspace.bronze.raw_locales) AS LOCALES;
```
