### a. Mejores sucursales

Top-10 de sucursales según monto vendido ordenado de mayor a menor. Debe tener ambas columnas.



```python
%sql
SELECT y.id_sucursal, y.nombre_sucursal, sum(x.monto_total) as monto_total_V

from workspace.silver.dim_vendedor as v
inner join workspace.gold.fact_ventas_final as x
on v.Id_vendedor=x.id_vendedor
inner join workspace.silver.dim_sucursal as y
on y.nombre_sucursal=v.sucursal_nombre
group by y.nombre_sucursal, y.id_sucursal
ORDER BY monto_total_V DESC
LIMIT 10
```

```python
%sql

SELECT s.id_sucursal, s.nombre_sucursal, SUM(f.monto_total) AS monto_vendido
FROM workspace.gold.fact_ventas_final f
INNER JOIN workspace.silver.dim_vendedor v
ON f.id_vendedor = v.id_vendedor
INNER JOIN workspace.silver.dim_sucursal s
ON v.sucursal_nombre = s.nombre_sucursal
GROUP BY s.id_sucursal, s.nombre_sucursal
ORDER BY monto_vendido DESC
LIMIT 10;
```

###b. Mejores vendedores con detalle de top 3 de productos

- Listado ordenado de mayor a menor con los mejores vendedores por “monto_total”.

- Agregar los 3 productos más vendidos por cada uno con sus cantidades correspondientes.


```python
%sql
with monto_vend as (select v.vendedor_nombre, sum(f.monto_total) as monto_total, f.id_producto, sum(f.cantidad) as cantidad from workspace.gold.fact_ventas_final as f 
inner join workspace.silver.dim_vendedor as v
on f.id_vendedor=v.Id_vendedor
group by v.vendedor_nombre, f.id_producto
order by monto_total DESC ),
rank as (
    select mv.vendedor_nombre, mv.monto_total, mv.id_producto, mv.cantidad, rank() over (partition by mv.vendedor_nombre order by mv.monto_total DESC) as ranking from monto_vend as mv
)
select * from rank where ranking between 1 and 3
```

### c. Peores sucursales

Peores sucursales: identificar las sucursales con monto vendido menor a $4.000.000.

```python
%sql
select v.sucursal_nombre as sucursal, sum(f.monto_total) as monto_total
from workspace.silver.dim_vendedor as v
inner join workspace.gold.fact_ventas_final as f
on v.Id_vendedor=f.id_vendedor 
group by v.sucursal_nombre
having sum(f.monto_total)<4000000
order by monto_total desc
```
